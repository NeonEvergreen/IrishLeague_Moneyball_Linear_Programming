```python
def visualise_metric_threshold(metric_distribution : pd.Series, threshold_, title : str)-> plt:
    plt.figure(figsize=(8,5),dpi=100)
    sns.histplot(metric_distribution, bins=10, kde=True)
    plt.plot(threshold_,0, marker="x", color="purple", markersize=15, clip_on=False)
    plt.title(label=title, fontweight=700, fontsize=12)
    plt.xticks(rotation=0, fontsize=15, fontweight=400)
    score_text_ = f"{round(threshold_,2)}({round(percentileofscore(metric_distribution,threshold_),1)}%)"
    plt.annotate(text = score_text_,
               xy= (threshold_,0),
               fontweight=200,
               color="red",
               xytext = (threshold_ - np.std(metric_distribution)/3,20),
               arrowprops={"color": "red", "lw" : 0.1})
    return plt


def visualise_boxplot(axes : plt.subplot , data : pd.DataFrame, y : str, hue : str, props : dict)-> plt.subplot:
    axes.set_title(props.get("title", ""), fontsize=13, fontweight=600)
    axes.set_xlabel(props.get("xlabel", ""), fontsize = 10)
    axes.set_ylabel(props.get("ylabel", ""), fontsize = 10)
    sns.boxplot(data = data ,
                y = y, 
                hue=hue,
               ax=axes,
               **props.get("visualisation", {}))
    return axes
    
```

# Linear Programming for Moneyball Team



## Linear Programming Parts

__This Linear Programming excerise builds a team out of the 2025 football League data that meets the rigorous performance metric requirements while minimising player acquisition costs__

* We treat all the players presence as the `Decision Variables`
* Ideally Linear programming assumes continuous `Decision Variables`, but we will treat them as probable team member scores bounded between $0$ and $1$
$x_{i} \in [0,1]$ refers to a players "part of the team" score or probability where,
    * $0$ is least likely
    * $1$ is most likely
    * $0 < x_{i} < 1$ treated as the `in the team score`
* The core objective of the Algorithm is to meet the `performance constraints` while minimising the `Acquisition costs` of players
    * The `performance constraints` are here divided into `Attack scores` $AP$ and `Defense scores` $DP$
    * Both metrics are built for each player using `Historical data`


---

### Objective Function

$$\text{Objective Function} = \text{Team Acquisition Cost} = \sum_{i=0}^{N}C_ix_i = C_1.x_1 + C_2.x_2 + \dots + C_N.x_N$$

$$\text{WHERE, } C_i \text{is the cost of the player "i" }$$

### Constraints

#### Attack Power constraint $\hspace{14.2cm}$ (TOP PERCENTILE [90% by default])

$$\text{Team ATTACK Power} \le \alpha_1.x_1 + \alpha_2.x_2 + \dots + \alpha_N.x_N$$

$$\text{WHERE, } \alpha_i \text{is the Attack power of the player "i" (Goals per 90 mins)}$$



#### Defence Power constraint $\hspace{14cm}$ (TOP PERCENTILE [90% by default])

$$\text{Team DEFENCE Power} \le \beta_1.x_1 + \beta_2.x_2 + \dots + \beta_N.x_N$$

$$\text{WHERE, } \beta_i \text{is the Defence power of the player "i" (Goals conceded per match weighted by role)}$$


#### TOTAL Player constraint $\hspace{14.4cm}$ (10 Players)

$$10 \ge x_1 + x_2 + \dots + x_N$$


---


### Matrix Representation

$$\text{MIN}_x\left(\begin{pmatrix}C_1&C_2&\dots&C_N\end{pmatrix}.\begin{pmatrix}x_1\\x_2\\\vdots\\x_N\end{pmatrix}\right)$$
Such that, $$\begin{pmatrix}\alpha_1&\alpha_2&\dots&\alpha_N\\\beta_1&\beta_2&\dots&\beta_N\\1&1&\dots&1\end{pmatrix}.\begin{pmatrix}x_1\\x_2\\\vdots\\x_N\end{pmatrix}\le\begin{pmatrix}\text{Attack Power}\\\text{Defence Power}\\\text{10}\end{pmatrix}$$

__NOTE__ : Scipy linprog algorithm solves only for bounded constraints, Since the actual need is to exceed the threshold for $AP$ and $DP$, we turn it into a bounded inequality by adding a $-$ to both sides


### Variables

* $x_i$ : Team suitability value for player $i$ ; $x\in[0,1]$  (`DECISION VARIABLES`})
* $C_i$ : Cost of player $i$
* $\alpha_i$ : Player $i$ Attack power
* $\beta_i$ :  Player $i$ Defence power

```python
def simplex(objective_coeff : list, constraint_coeff : np.array , constraint_values : list, silent : bool=False) -> None:
    bounds = [(0,1) for _ in range(len(objective_coeff))]

    res = linprog(objective_coeff, A_ub=constraint_coeff, b_ub=constraint_values,
                  bounds=bounds, method="simplex")

    if(not silent):
        print("\n")
        print(f"SOLVER STATUS : {res.message}")
        print(f"OPTIMAL OBJECTIVE VALUES : {res.fun}")
        # print(f"Optimum decision variables : {res.x}")
        print(f"SLACK : {res.slack}")
        print("\n")

    return  res

def build_constraints(df : pd.DataFrame, constraint_columns : List[list], constraint_coeff : list)-> Tuple[list, list]:
    assert(len(constraint_columns) == len(constraint_coeff))
    ## CONSTRAINTS
    constraint_list = list()
    for col in constraint_columns:
        constraint_coeff = df.loc[:, col]
        bounded_constraint_coeff = constraint_coeff.to_numpy() * -1
        constraint_list.append(bounded_constraint_coeff)
        
    ## Wrapping values
    constraints_coeff = [attack_coeff, defense_coeff, count_coeff]
    min_attack_req = -AP_limit
    min_defense_req = -DP_limit
    performance_limits = [min_attack_req, min_defense_req, selection_count]


def print_equations(constraints_coeff : np.array, objective_coeff : np.array, constraint_values : np.array)->None:
  for idx, constraint in enumerate(constraints_coeff):
    print(f"constraint {idx} : {constraint} <= {constraint_values[idx]}")
  print(f"Objective : {objective_coeff}")


def selected_players_df(sample_df : pd.DataFrame, res_x : dict, percentile : float = PERCENTILE, silent:bool=False) -> pd.DataFrame:
    sample_df["lin_prog_score"] = res_x
    selected_players = sample_df[sample_df["lin_prog_score"] > 0]

    if(not silent):
        print(f"{round(percentile * 100,2)}% AP : {round(AP_limit,6)}")
        print(f"{round(percentile * 100,2)}% DP : {round(DP_limit,6)}")
        print("Selected Team AP : ", np.sum(selected_players.loc[:,"gls_90"].to_numpy()))
        print("Selected Team DP : ", np.sum(selected_players.loc[:,"defence_metric"].to_numpy()))
        print("Selected Team Cost : €", np.sum(selected_players.loc[:,"player_cost_k"].to_numpy()))

    return selected_players


```

```python
############# SETTING PERFORMANCE LIMITS
# Finding the top percentile times average ATTACKER & DEFENDER Proportions

@interact(PERCENTILE=widgets.FloatSlider(min=0.1, max=0.99, step=0.01))
def main(PERCENTILE):    
    global AP_limit
    global DP_limit
    
    AP_limit = attackers_metric_distribution.quantile(q=PERCENTILE) * 4 #4 top % Attackers worth of Gls/90
    DP_limit = defenders_metric_distribution.quantile(q=PERCENTILE) * 4 #4 top % Defenders worth of Gls saved/90
    
    print(f"AP limit : {AP_limit}")
    print(f"DP limit : {DP_limit}")
    
    ############ LINEAR PROGRAMMING
    
    Number_of_players = df_complete_data.shape[0]
    selection_count = 10
    
    sample_players_df = df_complete_data.sample(n=Number_of_players,
                                                random_state=SEED,
                                             replace=False)
    
    ## CONSTRAINTS
    ones_ = np.full(shape=[1,Number_of_players],fill_value=1)

    #### Getting the coefficients of defense score
    defenders_metric_distribution_s = sample_players_df.loc[:, "defence_metric"]
    defense_coeff = defenders_metric_distribution_s.to_numpy() * -1
    #### Getting the coefficients for the Attack score
    attackers_metric_distribution_s = sample_players_df.loc[:, "gls_90"]
    attack_coeff = attackers_metric_distribution_s.to_numpy() * -1
    #### Player count coefficient
    count_coeff = ones_.flatten().tolist()

    ## OBJECTIVE FUNCTION
    cost_coeff = sample_players_df["player_cost_k"]

    ## Wrapping values
    constraints_coeff = [attack_coeff, defense_coeff, count_coeff]
    min_attack_req = -AP_limit
    min_defense_req = -DP_limit
    performance_limits = [min_attack_req, min_defense_req, selection_count]
    
    # print_equations(constraints_coeff, cost_coeff, performance_limits)
     
    res_x = simplex(objective_coeff = cost_coeff,  constraint_coeff=constraints_coeff,
            constraint_values = performance_limits).x
    
    return selected_players_df(sample_players_df, res_x, percentile=PERCENTILE)
```

# Research and Comparison with Other Algorithms

## Time Complexity and Algorithm Comparison (Computationally heavy [~14-18 mins runtime])

```python
import timeit
```

```python
def get_contraints(df : pd.DataFrame , sample_size : int = 10, seed : int = 42) -> tuple:
    '''
    returns [CONSTRAINT COEFF, CONSTRAINT THRESHOLD, OBJECTIVE COEFF]
    We create this function so as to utilise it to dynamically create constraints with differen sample sizes
    seed is maintained for same sample sizes (in order to make sure comparisons are valid and fair)
    '''
    AP_limit = attackers_metric_distribution.quantile(q=PERCENTILE) * 4 #4 top % Attackers worth of Gls/90
    DP_limit = defenders_metric_distribution.quantile(q=PERCENTILE) * 4 #4 top % Defenders worth of Gls saved/90
    
    ############ LINEAR PROGRAMMING
    
    Number_of_players = sample_size
    selection_count = 10
    
    sample_players_df = df.sample(n=Number_of_players,
                                                random_state=seed,
                                             replace=False)
    
    ## CONSTRAINTS
    ones_ = np.full(shape=[1,Number_of_players],fill_value=1)
    
    #### Getting the coefficients of defense score
    defenders_metric_distribution_s = sample_players_df.loc[:, "defence_metric"]
    defense_coeff = defenders_metric_distribution_s.to_numpy() * -1
    #### Getting the coefficients for the Attack score
    attackers_metric_distribution_s = sample_players_df.loc[:, "gls_90"]
    attack_coeff = attackers_metric_distribution_s.to_numpy() * -1
    #### Player count coefficient
    count_coeff = ones_.flatten().tolist()
    
    ## OBJECTIVE FUNCTION
    cost_coeff = sample_players_df["player_cost_k"]
    cost_coeff = cost_coeff.tolist()
    
    ## Wrapping values
    constraints_coeff = [attack_coeff, defense_coeff, count_coeff]
    constraints_coeff = [arr.tolist() if isinstance(arr, np.ndarray) else arr for arr in constraints_coeff]
    min_attack_req = -AP_limit
    min_defense_req = -DP_limit
    performance_limits = [min_attack_req, min_defense_req, selection_count]
    performance_limits = [float(val) for val in performance_limits]

    return (constraints_coeff, performance_limits, cost_coeff)


def visualise_performance(x : iter, y : iter, ax : plt.subplot, properties : dict)->plt.subplot:
    ''' 
    Function to visualise time required per run for different linprog Algorithms
    '''
    # a dynamic allocation for props.
    defaults = {"title" : "",
               "xlabel" : "",
               "ylabel" : "",
               "scatter_props" : dict(s=100),
               "line_props" : dict(color="red")}
    defaults.update(properties)
    sns.scatterplot(x=x,
                y=y,
                ax=ax,
                **defaults["scatter_props"])
    sns.lineplot(x=x, 
                 y=y,
                ax=ax,
                **defaults["line_props"])
    
    ax.set_title(defaults["title"], fontsize=12, fontweight=500)
    ax.set_xlabel(defaults["xlabel"], fontsize=10)
    ax.set_ylabel(defaults["ylabel"], fontsize=10)
    return ax
```

#### HiGHS vs Simplex (Robust Presolver handling)

```python
decision_variables_list = [3,5,10,15,20,30,50,100,200] # Number of decision variables. we create a range from 3 to 200

fig, axes = plt.subplots(nrows=3,ncols=2, figsize=(14,10))

## Algorithms (To capture time taken for each code)
algorithm_type = {"simplex" : list(),  "interior-point" : list(), "revised simplex" : list(), "highs" : list(), "highs-ipm" : list(), "highs-ds" : list()}

for algo in algorithm_type:
    for n in decision_variables_list:
        constraints = get_contraints(df_complete_data, n)

        ## The setup code (excludes our import statement and constraint assignment from code implementation)
        setup_code = f"""from scipy.optimize import linprog
objective = {constraints[2]}
constraint_coeff = {constraints[0]}
constraint_thresh = {constraints[1]}
        """

        # We feed code as string, the eval required adds overhead. But since we are doing relative comparisons this should be uniform
        simplex_with_N_decision_vars = f'''linprog(objective, A_ub=constraint_coeff, b_ub=constraint_thresh,bounds=(0,1), method="{algo}")'''
        algorithm_type[algo].append(timeit.timeit(simplex_with_N_decision_vars, setup=setup_code, number = 1000)/1000)



### Visualising time Performance based on algorithm type (Simplex, IPM, revised Simplex VS HiGHS (Revised Simplex, IPM, Dual Simplex) which has advanced presolver routines)

visualise_performance(x= decision_variables_list,
                      y=algorithm_type["simplex"],
                      ax = axes[0][0],
                      properties=dict(title = "Simplex",
                                     xlabel = "Decision variable count",
                                     ylabel = "Time taken (seconds)")
                     )


visualise_performance(x= decision_variables_list,
                      y=algorithm_type["highs"],
                      ax = axes[0][1],
                      properties=dict(title = "HiGHS (revised Simplex)",
                                     xlabel = "Decision variable count",
                                     ylabel = "Time taken (seconds)")
                     )


visualise_performance(x= decision_variables_list,
                      y=algorithm_type["interior-point"],
                      ax = axes[1][0],
                      properties=dict(title = "Interior Point Method",
                                     xlabel = "Decision variable count",
                                     ylabel = "Time taken (seconds)")
                     )


visualise_performance(x= decision_variables_list,
                      y=algorithm_type["highs-ipm"],
                      ax = axes[1][1],
                      properties=dict(title = "HiGHS (Interior point method)",
                                     xlabel = "Decision variable count",
                                     ylabel = "Time taken (seconds)")
                     )



visualise_performance(x= decision_variables_list,
                      y=algorithm_type["revised simplex"],
                      ax = axes[2][0],
                      properties=dict(title = "Revised Simplex",
                                     xlabel = "Decision variable count",
                                     ylabel = "Time taken (seconds)")
                     )


visualise_performance(x= decision_variables_list,
                      y=algorithm_type["highs-ds"],
                      ax = axes[2][1],
                      properties=dict(title = "HiGHS (Dual Simplex)",
                                     xlabel = "Decision variable count",
                                     ylabel = "Time taken (seconds)")
                     )


fig.suptitle("Optimisation time vs Number of decision variables", fontsize=14, fontweight=700)
plt.tight_layout()
```

## Conclusion

A relative comparison of different __Linear programming__ algorithms, including with and without strong `Presolver` and `Postsolver` routines.

The current research for efficiency on `Linear programming` focuses extensively on these aforementioned `presolver` and `postsolver` operations along with `Interior Point Methods` and their variants, which reduce the `iterations` (number of steps required) towards the optimal point.

These steps focus on two aspects of `Linear Programming`

    * Efficiency of computation
    * Ability to solve a problem
    
The most advanced `algorithm` open-source right now is thr `HiGHS`, but commercial solvers like `MOSEK` and especially `COPT` show phenomenally better performances with their propietary dynamic `presolver`s and `postsolver`s

   * `COPT` is the leading algorithm, being able to solve at an amazing $17.9$ times Faster than the average speed of `HiGHS` ($17.9 \times$)
   * `COPT` also solved $25%$ more `LP` problems that `HiGHS` was unable to solve
    
A few areas, research currently focus on are :
    * `facial reduction`
    * `anti cycling algorithms`
    * `numerical stability` through `iteration precision`s
    * `Duality`
    * `Scaling`
    * `Pivoting rules`

* Of these, `Scaling` was attempted, but results didnt show a difference.
* Moreover `Scaling` is often problem dependent on performance, along with more influential on `constraint` equalities
* The `constraints` of the Moneyball challenge were already uniform in scales (apart from a few outliers), which supported the previous result.
